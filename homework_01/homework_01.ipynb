## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**

- `23456 -> 25436`
- `30789 -> 38709`


```python
number = int(input())

if 10000 <= number <= 99999:
    number_string = str(number)
    reversed = number_string[-2:0:-1]

    result = int(f"{number_string[0]}{reversed}{number_string[-1]}")

    print(result)
else:
    print("Ожидалось пятизначное число")
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.

<br>

**Примеры:**

- `4  -> 0`
- `6  -> 1`
- `14 -> 4`


```python
days_till_vacation = int(input())

weekends_till_vacation = 2 * (days_till_vacation // 7)
if (days_till_vacation % 7) > 5:
    weekends_till_vacation += 1

print(weekends_till_vacation)
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**

- `3, 4, 6  -> True`
- `5, 7, 8  -> False`
- `4, 5, 12 -> True`


```python
width = int(input("Ширина плитки:"))
height = int(input("Высота плитки:"))
required_pieces = int(input("Сколько нужно долек:"))

# Поскольку требуется совершить разлом то проверка строгая
is_horizontal_possible = required_pieces % width == 0 and required_pieces / width < height
is_vertical_possible = required_pieces % height == 0 and required_pieces / height < width

break_is_possible = is_horizontal_possible or is_vertical_possible
print(break_is_possible)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**

- `3   -> III`
- `15  -> XV`
- `234 -> CCXXXIV`


```python
def roman_group(result: str, number: int, module: int, group: str):
    value, number = divmod(number, module)
    if value == 9:
        result += group[::2]
    elif value >= 5:
        result += group[1] + (value - 5) * group[0]
    elif value == 4:
        result += group[0:2]
    else:
        result += value * group[0]

    return (result, number)


number = int(input("Введите число:"))
if number > 3999:
    print("Число слишком большое (>3999)")
elif number < 0:
    print("Число отрицательное")
else:
    result = ""

    result, number = roman_group(result, number, 1000, "M--")
    result, number = roman_group(result, number, 100, "CDM")
    result, number = roman_group(result, number, 10, "XLC")
    result, number = roman_group(result, number, 1, "IVX")

    if not result:
        result = "Nulla"

    print(result)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**

- `5.6  -> True`
- `.78  -> True`
- `.67. -> False`
- `5    -> True`


```python
number = input()

is_negative = False
is_number = False

if len(number) > 0:
    if number[0] == "+":
        number = number[1:]
    elif number[0] == "-":
        is_negative = True
        number = number[1:]

    digits = 0
    is_float = False
    is_number = True
    for ch in number:
        if ch == ".":
            if is_float:
                # Больше одной запятой
                is_number = False
                break

            is_float = True
        elif "0" <= ch <= "9":
            # Цифра
            digits += 1
        else:
            # Не цифра и не запятая
            is_number = False
            break

    if digits == 0:
        is_number = False

if is_number:
    print(f"Вещественное: {is_number}, Отрицательное: {is_negative}")
else:
    print(f"Вещественное: {is_number}")
```
